```python
# Load audio file
y, sr = librosa.load('devided_data/Train/blues/blues.00009.wav')

# Extract features
mel_spectrogram = librosa.feature.melspectrogram(y=y, sr=sr)
chroma = librosa.feature.chroma_stft(y=y, sr=sr)
mfcc = librosa.feature.mfcc(y=y, sr=sr)

# Normalize features
mel_norm = librosa.util.normalize(mel_spectrogram)
chroma_norm = librosa.util.normalize(chroma)
mfcc_norm = librosa.util.normalize(mfcc)

# Concatenate features
features_concatenated = np.concatenate((mel_norm, chroma_norm, mfcc_norm), axis=0)

# Plot and save concatenated features as an image
plt.figure(figsize=(10, 6))
librosa.display.specshow(features_concatenated, y_axis='mel', fmax=8000, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Concatenated Audio Features')
plt.tight_layout()
#plt.savefig('concatenated_features.png')
```

```python

from tensorflow.keras.models import Sequential, Model
from tensorflow.keras import layers, optimizers, callbacks
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from tensorflow.keras.models import Sequential, Model
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing.image import ImageDataGenerator
from keras.callbacks import ModelCheckpoint, ReduceLROnPlateau, EarlyStopping
from tensorflow.keras.utils import image_dataset_from_directory
```

```python
# Define the data directories
train_dir = 'images_data/Train'
val_dir = 'images_data/Validation'
test_dir = 'images_data/Test'
```

```python
from tensorflow.keras.preprocessing.image import array_to_img, img_to_array, load_img

img = load_img('images_data/Train/blues/blues.00001.png')
x = img_to_array(img)/255  # We divide the pixel value by 255 to end up between 0 and 1
x.shape
```

```python

model = Sequential()

model.add(layers.Rescaling(1./255, input_shape = (173, 1293, 3)))


model.add(layers.Conv2D(filters = 32, kernel_size = (3,3), activation="relu", padding = "same"))
model.add(layers.MaxPooling2D(pool_size=(2, 2), padding = "same") )


model.add(layers.Conv2D(filters = 64, kernel_size = (3,3), activation="relu", padding = "same"))
model.add(layers.MaxPooling2D(pool_size=(2, 2), padding = "same") )


model.add(layers.Conv2D(filters = 128, kernel_size = (3,3), activation="relu", padding = "same"))
model.add(layers.MaxPooling2D(pool_size=(2, 2), padding = "same") )

model.add(layers.Conv2D(filters = 256, kernel_size = (3,3), activation="relu", padding = "same"))
model.add(layers.MaxPooling2D(pool_size=(2, 2), padding = "same") )

model.add(layers.Flatten())

# Here we flatten our data to end up with just one dimension

model.add(layers.Dense(128, activation="relu"))

model.add(layers.Dropout(0.5))

model.add(layers.Dense(10, activation="softmax"))
```

```python

adam = optimizers.Adam(learning_rate = 0.001)
model.compile(loss='categorical_crossentropy',
              optimizer= adam,
              metrics=['accuracy'])
```

```python
batch_size = 32
```

```python
train_ds = image_dataset_from_directory(
  train_dir, # Folder directory 
  labels = "inferred", # inferred from
  label_mode = "categorical",
  seed=123,
  image_size=(173, 1293),
  batch_size=batch_size)

# We define a second one for the test data

val_ds = image_dataset_from_directory(
  val_dir,
  labels = "inferred",
  label_mode = "categorical",
  seed=123,
  image_size=(173, 1293),
  batch_size=batch_size)
```

```python
class_names = train_ds.class_names
print(class_names)
```

```python


MODEL = "model_1"

#This callback will save the model to a file after every epoch.
modelCheckpooint = callbacks.ModelCheckpoint("{}.keras".format(MODEL), monitor="val_loss", verbose=0, save_best_only=True)

#This callback reduces the learning rate when a metric has stopped improving.
LRreducer = callbacks.ReduceLROnPlateau(monitor="val_loss", factor = 0.1, patience=3, verbose=1, min_lr=0)

#This callback will stop the training if the monitored metric (in this case, validation loss) does not improve.
EarlyStopper = callbacks.EarlyStopping(monitor='val_loss', patience=10, verbose=0, restore_best_weights=True)
```

```python
%%time
history = model.fit(
        train_ds,
        epochs=60,
        validation_data=val_ds,
        callbacks = [modelCheckpooint, LRreducer, EarlyStopper])
```

```python
# We'll use a model with the same convolutional layers, but we'll add Augmentation layers before that

model_2 = Sequential()

model_2.add(layers.Rescaling(1./255, input_shape = (150, 150, 3)))

# Data Augmentation Layers

model_2.add(layers.RandomFlip("horizontal"))
model_2.add(layers.RandomZoom(0.1))
model_2.add(layers.RandomTranslation(0.2, 0.2))
model_2.add(layers.RandomRotation(0.1))


# Convolutional Layers

model_2.add(layers.Conv2D(filters = 32, kernel_size = (3,3), activation="relu", padding = "same"))
model_2.add(layers.MaxPooling2D(pool_size=(2, 2), padding = "same") )


model_2.add(layers.Conv2D(filters = 32, kernel_size = (3,3), input_shape = (150, 150, 3), activation="relu", padding = "same"))
model_2.add(layers.MaxPooling2D(pool_size=(2, 2), padding = "same") )


model_2.add(layers.Conv2D(filters = 64, kernel_size = (3,3), input_shape = (150, 150, 3), activation="relu", padding = "same"))
model_2.add(layers.MaxPooling2D(pool_size=(2, 2), padding = "same") )

model_2.add(layers.Conv2D(filters = 128, kernel_size = (3,3), input_shape = (150, 150, 3), activation="relu", padding = "same"))
model_2.add(layers.MaxPooling2D(pool_size=(2, 2), padding = "same") )

model_2.add(layers.Flatten())

model_2.add(layers.Dense(64, activation="relu"))

model_2.add(layers.Dropout(0.5))

model_2.add(layers.Dense(10, activation="softmax"))
```

```python
adam = optimizers.Adam(learning_rate = 0.001)

model_2.compile(loss='categorical_crossentropy',
              optimizer= adam,
              metrics=['accuracy'])
```

```python
MODEL = "model_2"

modelCheckpoint = callbacks.ModelCheckpoint("{}.keras".format(MODEL), monitor="val_loss", verbose=0, save_best_only=True)

LRreducer = callbacks.ReduceLROnPlateau(monitor="val_loss", factor = 0.1, patience=3, verbose=1, min_lr=0)

EarlyStopper = callbacks.EarlyStopping(monitor='val_loss', patience=10, verbose=0, restore_best_weights=True)
```

```python
%%time
history_2 = model_2.fit(
        train_ds,
        epochs=30,
        validation_data=val_ds,
        callbacks = [modelCheckpoint, LRreducer, EarlyStopper]
        )
```

```python
data_augmentation = Sequential()

data_augmentation.add(layers.RandomFlip("horizontal"))
data_augmentation.add(layers.RandomZoom(0.1))
data_augmentation.add(layers.RandomTranslation(0.2, 0.2))
data_augmentation.add(layers.RandomRotation(0.1))
```

```python
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input
from tensorflow.keras.models import Model
```

```python
base_model = VGG16(weights = "imagenet", include_top = False, input_shape = (150, 150, 3))
```

```python
# First step is to initialize the VGG16 model but without the top as we'll adapt it to our problem


inputs = layers.Input(shape = (150, 150, 3))

x = data_augmentation(inputs) # We still have our data augmentation layers
x = preprocess_input(x) # Then a preprocessing layer specifically designed for the VGG16
x = base_model(x) # Then our transfer learning model

x = layers.Flatten()(x) # Followed by our custom dense layers, tailored to our binary task

x = layers.Dense(128, activation = "relu")(x)
x = layers.Dropout(0.3)(x)
x = layers.Dense(64, activation = "relu")(x)
x = layers.Dropout(0.3)(x)
pred = layers.Dense(10, activation = "softmax")(x)

# We use the keras Functional API to create our keras model

model_3 = Model(inputs = inputs , outputs = pred)

# And we freeze the VGG16 model

base_model.trainable = False

#for layer in base_model.layers[:10]:
#  layer.trainable = False

```

```python
adam = optimizers.Adam(learning_rate = 0.001)
model_3.compile(loss='categorical_crossentropy',
              optimizer=adam,
              metrics=['accuracy'])
```

```python
MODEL = "model_3"

modelCheckpoint = callbacks.ModelCheckpoint("{}.h5".format(MODEL), monitor="val_loss", verbose=0, save_best_only=True)

LRreducer = callbacks.ReduceLROnPlateau(monitor="val_loss", factor = 0.1, patience=3, verbose=1, min_lr=0)

EarlyStopper = callbacks.EarlyStopping(monitor='val_loss', patience=10, verbose=0, restore_best_weights=True)
```

```python
%%time
history_3 = model_3.fit(
        train_ds,
        epochs=30,
        validation_data=val_ds,
        callbacks = [modelCheckpoint, LRreducer, EarlyStopper])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, BatchNormalization, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau, EarlyStopping
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define the model
model = Sequential()

# Rescaling
model.add(Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(173, 1293, 3)))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2), padding='same'))

model.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2), padding='same'))

model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2), padding='same'))

model.add(Conv2D(256, (3, 3), activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2), padding='same'))

model.add(Conv2D(512, (3, 3), activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2), padding='same'))

model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))

# Compile the model
adam = Adam(learning_rate=0.001)
model.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy'])

# Data Augmentation
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')

val_datagen = ImageDataGenerator(rescale=1./255)

train_ds = train_datagen.flow_from_directory(
    train_dir,
    target_size=(173, 1293),
    batch_size=batch_size,
    class_mode='categorical')

val_ds = val_datagen.flow_from_directory(
    val_dir,
    target_size=(173, 1293),
    batch_size=batch_size,
    class_mode='categorical')

# Callbacks
MODEL = "model_1"
modelCheckpoint = ModelCheckpoint(f"{MODEL}.keras", monitor="val_loss", verbose=0, save_best_only=True)
LRreducer = ReduceLROnPlateau(monitor="val_loss", factor=0.1, patience=3, verbose=1, min_lr=0.00001)
EarlyStopper = EarlyStopping(monitor='val_loss', patience=10, verbose=0, restore_best_weights=True)

# Train the model
history = model.fit(
    train_ds,
    epochs=100,
    validation_data=val_ds,
    callbacks=[modelCheckpoint, LRreducer, EarlyStopper])
```
